```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

toronto_onehot
```

```python
# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns =  list(toronto_onehot.loc[:, toronto_onehot.columns == 'Neighborhood']) + list(toronto_onehot.loc[:, toronto_onehot.columns != 'Neighborhood'])

toronto_onehot=toronto_onehot[fixed_columns]
toronto_onehot.head()
```

```python
toronto_grouped=toronto_onehot.groupby('Neighborhood').mean().reset_index()

toronto_grouped
```

```python
from sklearn.cluster import KMeans
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
toronto_grouped.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python

toronto_merged = toronto_grouped.merge(toronto_venues,how='inner',left_on='Neighborhood',right_on='Neighborhood').groupby('Neighborhood').max().reset_index()
toronto_grouped
```

For this exercise I followed the same method as the previous one but applied to Toronto instead New York. I took the top 100 venues per Borough from Four Square, and then used one hot encoding based on the Venue Type. I then clustered this (using 5 clusters) and displayed these clusters on a map. I found a dominant cluster concenrated the centre, with a secondary cluster in the ring outside the centre of the city. There were 3 smaller clusters scattered in the suburbs/edge of the city. 

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[2:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[1:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+2, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+2))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_merged.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 2:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head(100)
```
